# Propagating and non propagating waves

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
from scipy import integrate
from numpy.linalg import inv
%matplotlib notebook

f = 1
w = f*np.pi*2
c = 1
k = 40
T = 1

x = np.linspace(0,20,500)
t = np.linspace(0,20,100)

gamma = pow(w**2/c**2-k/T,0.5)
print(gamma)


fig=plt.figure()
ax = plt.axes(xlim=(0,20),ylim=(-2, 2),xlabel='position, x',ylabel = 'wave response')
ax.grid()
line, = ax.plot([], [], lw=2,color='r')

def init():
    line.set_data([], [])
    return line

def animate(i):
    y = np.exp(1j*(gamma*x-w*t[i]))
    line.set_data(x, y.real)
    return line

anim = FuncAnimation(fig, animate, init_func=init,frames=1000, interval=500, blit=True)
```
